Correlation Heatmap

```python
corr = df.select_dtypes(['int64', 'float64']).corr()
sns.heatmap(corr, annot=True, cmap='YlGnBu', center=0)
plt.title('Correlation Heatmap')
```

```python
cov = df.select_dtypes(['int64', 'float64']).cov()
sns.heatmap(cov, annot=True, cmap='coolwarm', center=0)
plt.title('Covariance Heatmap')
```
