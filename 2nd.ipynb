Hypothesis Testing


```python
s_x = 90
p_U = 82
p_std = 20
n = 81
ap = 0.05
```

```python
z_cal = (s_x - p_U) / (p_std / np.sqrt(n))
print(f"Calculated Z-value: {z_cal}")
z_cal
```

```python
z_table = stats.norm.ppf(1 - ap)
print(f"Z-table value for 0.95: {z_table}")
```

```python
if z_table < z_cal:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
print("using Z-test for hypothesis testing of first type")
```

```python
old_data = np.random.randint(1, 100, size=30)
old_data
np.mean(old_data)
```

```python
new_data = np.random.randint(1, 100, size=30)
new_data
np.mean(new_data)
```

```python
pop_std = 2.5
n_sp = len(new_data)
apl = 0.05
z_table = stats.norm.ppf(1 - ap)
z_table
```

```python
z_cal = (np.mean(new_data) - np.mean(old_data)) / (pop_std / np.sqrt(n_sp))
print(f"Calculated Z-value for new data: {z_cal}")
if z_table < z_cal:
    print("Reject the null hypothesis for new data")
else:
    print("Fail to reject the null hypothesis for new data")
```
