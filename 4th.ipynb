Feature Scaling Techniques
In this section, we will discuss various techniques for feature scaling, which is a crucial step in many
machine learning algorithms.
### Scaling Techniques
There are several techniques for scaling features, including:
*   **Standardization**: This involves subtracting the mean and dividing by the standard deviation for each
feature. This is also known as Z-scoring.
*   **Normalization**: This involves scaling features to a common range, usually between 0 and
1. This is also known as Min-Max scaling.
*   **Log Scaling**: This involves taking the logarithm of the features. This is useful for
features that have a large range of values.
*   **Robust Scaling**: This involves scaling features based on the interquartile range (
    IQR
    ) rather than the standard deviation. This is useful for features that have outliers.
    ### Scaling
    Scaling is a crucial step in many machine learning algorithms. It helps to prevent features with large
    ranges from dominating the model. There are several techniques for scaling features, including standardization,
    normalization, log scaling, and robust scaling.

## Function Transform :
is a technique used to convert data from one format or structure to another. This is often necessary when preparing data for analysis or modeling, as different algorithms may require specific input formats.

```python
sns.displot(dataset["CoapplicantIncome"], kde=True, color="purple")   
plt.show()
```

```python
q1 = dataset["CoapplicantIncome"].quantile(0.25)
q3 = dataset["CoapplicantIncome"].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
```

# Feature Selection Technique
The feature selection technique is used to select the most relevant features from the dataset. The goal is to
reduce the dimensionality of the data while preserving the most important information. In this study, we
used the Recursive Feature Elimination (RFE) technique, which is a popular feature selection method.
RFE is a wrapper method that recursively eliminates the least important features until a specified number
of features is reached.
### 1. Forward Elimination
is a feature selection method that removes the least significant features from the dataset one by one.
### 2. Backward Substitution
is a feature selection method that starts with all features and removes the least significant ones until the desired number of features is reached.

# Cost Function
used to minimize error
# 1. Mean Squared Error (MSE)
also known as L2 Loss
Fromula : (1/n) * sum((y_true - y_pred)^2)
# 2. Root Mean Squared Error (RMSE)
Formula : sqrt((1/n) * sum((y_true - y_pred)^2))
# 3. Mean Absolute Error (MAE)
Also known as L1 loss
Formula : (1/n) * sum(abs(y_true - y_pred))

```python
# cost function used in linear regression is the Mean Squared Error (MSE)
# for finding the best fit line, we minimize the MSE
# cost function is used to measure how well the model fits the data
# the goal is to find the parameters (slope and intercept) that minimize the cost function
# we use gradient descent to find the optimal parameters
# gradient descent is an optimization algorithm used to minimize the cost function
```

# Regularization Techniques
Regularization techniques are used to prevent overfitting by adding a penalty term to the loss function.
### L1 Regularization (Lasso)
L1 regularization adds a term to the loss function that is proportional to the absolute value of the model
parameters. This is also known as Lasso regression.
### L2 Regularization (Ridge)
L2 regularization adds a term to the loss function that is proportional to the square of the model parameters
This is also known as Ridge regression.
### Elastic Net Regularization
Elastic net regularization combines L1 and L2 regularization by adding a term proportional to the absolute valu
of the model parameters and a term proportional to the square of the model parameters.
