# 1. Logistic Regression
Logistic regression is a type of regression analysis that is used to predict the outcome of a categorical dependent
variable based on one or more predictor variables. It is a popular algorithm for classification problems, and it
is widely used in many fields, including marketing, finance, and healthcare.
### there is three types of logistic regression
1. Binary Logistic Regression: This is the most common type of logistic regression, where the dependent variabl
is a binary variable (0/1, yes/no, etc.). The goal is to predict the probability of the dependent variable being 1.
2. Multinomial Logistic Regression: This type of logistic regression is used when the dependent variable is a 
nominal variable with more than two categories. The goal is to predict the probability of each category.
3. Ordinal Logistic Regression: This type of logistic regression is used when the dependent variable has an order or 
ranking. The goal is to predict the probability of each category.

# Confusion Matrix
confusion matrix is used to evaluate the performance of a classification model. It is a table that displays the number of true positives
TP, true negatives TN, false positives FP, and false negatives FN. The confusion matrix is used
to calculate the accuracy, precision, recall, and F1 score of a model.


```python
false_positives = np.sum((y_test == 0) & (lr1.predict(X_test) == 1))
false_negatives = np.sum((y_test == 1) & (lr1.predict(X_test) == 0))
true_positives = np.sum((y_test == 1) & (lr1.predict(X_test) == 1))
true_negatives = np.sum((y_test == 0) & (lr1.predict(X_test) == 0))
print(f"False Positives: {false_positives}")
print(f"False Negatives: {false_negatives}")
print(f"True Positives: {true_positives}")
print(f"True Negatives: {true_negatives}")
```
